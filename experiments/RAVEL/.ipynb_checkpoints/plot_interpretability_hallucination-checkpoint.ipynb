```python
import sys

sys.path.append("../..")
%load_ext autoreload
%autoreload 2
```

```python
from datasets import load_from_disk
from torch.utils.data import DataLoader
from transformers import AutoTokenizer

from src.hyperdas.data_utils import (
    filter_dataset,
    generate_ravel_dataset,
    get_ravel_collate_fn,
)
```

```python
tokenizer = AutoTokenizer.from_pretrained("/scr-ssd/sjd24/llama3-8b")

tokenizer.padding_side = "left"
tokenizer.pad_token = tokenizer.eos_token
tokenizer.pad_token_id = tokenizer.eos_token_id
```

```python
from src.hyperdas.llama3.model import RavelInterpretorHypernetwork

hypernetwork = RavelInterpretorHypernetwork(
    model_name_or_path="/scr-ssd/sjd24/llama3-8b",
    num_editing_heads=32,
    intervention_layer=15,
    subspace_module="ReflectSelect",
    das_dimension=128,
    break_asymmetric=False,
)
hypernetwork = hypernetwork.to("cuda")
hypernetwork.load_model("/scr-ssd/sjd24/occupation/final_model")
```

```python
test_dataset = load_from_disk("../../experiments/RAVEL/data/occupation_test")

collate_fn = get_ravel_collate_fn(
    tokenizer,
    add_space_before_target=True,
    contain_entity_position=True,
    source_suffix_visibility=False,
    base_suffix_visibility=False,
)
dataloader = DataLoader(
    test_dataset, batch_size=16, collate_fn=collate_fn, shuffle=False
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=1.5)
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

hypernetwork.load_model("/scr-ssd/sjd24/occupation/final_model")

hypernetwork.plot_heatmap(
    dataloader,
    idxs=21,
    inference_mode=None,
    annot=False,
    batch_size=16,
    indicate_masked_tokens=False,
    digits=1,
    axes=axes[0],
)

hypernetwork.load_model("/scr-ssd/sjd24/occupation_shitty/final_model")

hypernetwork.plot_heatmap(
    dataloader,
    idxs=21,
    inference_mode=None,
    annot=False,
    batch_size=16,
    indicate_masked_tokens=False,
    digits=1,
    axes=axes[1],
)

hypernetwork.load_model("/scr-ssd/sjd24/occupation_shittier/final_model")

hypernetwork.plot_heatmap(
    dataloader,
    idxs=21,
    inference_mode=None,
    annot=False,
    batch_size=16,
    indicate_masked_tokens=False,
    digits=1,
    axes=axes[2],
)

instruction = axes[0].get_title().split("Label:")[0].strip()
axes[0].set_title("With Sparsity Loss", fontsize=20)
axes[1].set_title("With No Sparsity Loss", fontsize=20)
axes[2].set_title("With Too Much Sparsity Loss", fontsize=20)

axes[0].set_xticklabels(axes[0].get_xticklabels(), fontsize=10)
axes[1].set_xticklabels(axes[1].get_xticklabels(), fontsize=10)
axes[2].set_xticklabels(axes[2].get_xticklabels(), fontsize=10)

axes[0].set_ylabel("Counterfactual Sentence")
axes[1].set_ylabel("")
axes[2].set_ylabel("")

axes[1].set_yticklabels("")
axes[2].set_yticklabels("")

# Remove the colorbar for the first two plots
axes[0].collections[0].colorbar.remove()
axes[1].collections[0].colorbar.remove()


plt.subplots_adjust(wspace=0.1)

fig.savefig("sparsity_loss_comparison.pdf")
```
